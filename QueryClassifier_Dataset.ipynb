```python
import random
import pandas as pd

# Predefined templates for chit-chat and topic-based queries
chit_chat_templates = [
    "Hi, how are you?",
    "What's up?",
    "Tell me a joke.",
    "Do you like pizza?",
    "What's your favorite color?",
    "How's the weather?",
    "Do you have a hobby?",
    "Tell me something fun.",
    "What's the latest news?",
    "Do you watch movies?",
    "Have you been to any good restaurants lately?",
    "What music do you listen to?",
    "What's your favorite movie?",
    "Can we talk about something funny?",
    "What do you usually do for fun?",
    "Do you prefer tea or coffee?",
    "How's your day going?",
    "What's the most exciting thing you've done recently?",
    "Do you believe in aliens?",
    "What's your favorite holiday destination?"
]

topic_templates = [
    "What is the capital of {}?",
    "Explain the effects of {}.",
    "Who discovered {}?",
    "Define {}.",
    "What are the symptoms of {}?",
    "What is the boiling point of {}?",
    "Explain {}'s theory of relativity.",
    "What is the population of {}?",
    "What is the chemical formula for {}?",
    "Describe the history of {}.",
    "What are the causes of {}?",
    "What are the benefits of {}?",
    "How does {} impact our daily lives?",
    "What is the significance of {}?",
    "What is the future of {}?",
    "How can {} be prevented?",
    "What are the applications of {}?",
    "What are the risks associated with {}?",
    "What are the main challenges of {}?",
    "How does {} influence society?"
]

# Examples for placeholders in topic templates
topics = [
    "Health",
    "Environment",
    "Technology",
    "Economy",
    "Entertainment",
    "Sports",
    "Politics",
    "Education",
    "Travel",
    "Food"
]

# Function to generate chit-chat queries
def generate_chit_chat(n=500):
    return random.choices(chit_chat_templates, k=n)

# Function to generate topic-based queries
def generate_topic_queries(n=500):
    queries = []
    for _ in range(n):
        template = random.choice(topic_templates)
        topic = random.choice(topics)
        queries.append(template.format(topic))
    return queries

# Generate dataset
num_samples = 500  # Number of samples for each class
chit_chat_queries = generate_chit_chat(num_samples)
topic_queries = generate_topic_queries(num_samples)

# Combine data into a pandas DataFrame
data = {
    "query": chit_chat_queries + topic_queries,
    "label": [0] * num_samples + [1] * num_samples  # 0 for chit-chat, 1 for topic
}
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("query_dataset.csv", index=False)

# Display sample dataset
print(df.head())
```
